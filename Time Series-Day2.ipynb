```python
import pandas as pd
from datetime import datetime
import pandas_datareader as pdr
```

```python
df_tesla=pdr.get_data_yahoo('TSLA')
```

```python
df_tesla
```

```python
#Simple moving average
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open:10 days rolling']=df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
#Min_periods=1 karne se sirf first element hi NaN aayega
#Jabki pehle 9 elements NaN aana chahiye(As window=10)
df_tesla[['Open','Open:10 days rolling']].plot()
```

```python
df_tesla[['Open','Open:10 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

```python
#Note,we passed the data,in the form of a list
```

#We use this simple moving average,to know,when to buy or sell the stocks,eg ehrn the Blue line is higher than orange,hold it,and sell when blue line intersects and goes below orange line

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla[['Open','Open:30 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

```python
df_tesla['Open:50 days rolling']=df_tesla['Open'].rolling(window=50,min_periods=1).mean()
df_tesla[['Open','Open:10 days rolling','Open:30 days rolling','Open:50 days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

```python
##Expanding
##Cumulative Moving Average

df_tesla['Open'].expanding().mean().plot(figsize=(10,5))
```
